```python
cursor = conn.cursor()

for _, row in df.iterrows():

    sql = """
    INSERT INTO patients (
        age, sex, cp, trestbps, chol,
        fbs, restecg, thalach, exang,
        oldpeak, slope, ca, thal, target
    )
    VALUES (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)
    """

    values = (
        int(row['age']),
        int(row['sex']),
        int(row['cp']),
        int(row['trestbps']),
        int(row['chol']),
        int(row['fbs']),
        int(row['restecg']),
        int(row['thalach']),
        int(row['exang']),
        float(row['oldpeak']),
        int(row['slope']),
        int(row['ca']),
        int(row['thal']),
        int(row['target'])
    )

    cursor.execute(sql, values)

conn.commit()

print("All records inserted successfully!")
```
