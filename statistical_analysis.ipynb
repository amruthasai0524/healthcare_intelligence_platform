# Test 1: Independent T-Test

```python
disease = df[df['target']==1]['chol']

no_disease = df[df['target']==0]['chol']

t_stat, p_value = ttest_ind(
    disease,
    no_disease
)

print("T Statistic:", t_stat)
print("P Value:", p_value)
```

# Null Hypothesis (H₀)

## There is no significant difference in cholesterol levels.

# Test 2: Chi-Square Test

# Null Hypothesis (H₀)

## Gender and heart disease are independent.

# Alternative Hypothesis (H₁)

### Gender and heart disease are associated.

# Create Contingency Table

```python
table = pd.crosstab(
    df['sex'],
    df['target']
)

table
```

# Perform Test

```python
chi2, p, dof, expected = chi2_contingency(table)

print("Chi Square:", chi2)
print("P Value:", p)
```

# Interpretation

```python
if p < 0.05:
    print("Significant association")
else:
    print("No significant association")
```

# Interview Explantion

# Test 3: ANOVA

# Null Hypothesis (H₀)

* Mean age is same across all chest pain groups.

# Alternative Hypothesis (H₁)

* At least one group has a different mean age.

```python
group1 = df[df['cp']==0]['age']
group2 = df[df['cp']==1]['age']
group3 = df[df['cp']==2]['age']
group4 = df[df['cp']==3]['age']
```

```python
f_stat, p_value = f_oneway(
    group1,
    group2,
    group3,
    group4
)

print("F Statistic:", f_stat)
print("P Value:", p_value)
```

# Interpretation

```python
if p_value < 0.05:
    print("Significant difference")
else:
    print("No significant difference")
```

ANOVA was used to compare average age across multiple chest pain categories.

# 5. Confidence Interval

#Estimate population cholesterol range with 95% confidence.

```python
mean = df['chol'].mean()

std = df['chol'].std()

n = len(df)

margin = 1.96 * (std / np.sqrt(n))

lower = mean - margin

upper = mean + margin

print("95% Confidence Interval")
print(lower, upper)
```

# INterpretation 

#We are 95% confident that the true average cholesterol level lies between the calculated lower and upper bounds.

## Objective

The objective of this analysis was to statistically evaluate the relationship between clinical features and heart disease occurrence.

## Tests Performed

### 1. Independent T-Test

A T-Test was conducted to compare cholesterol levels between patients diagnosed with heart disease and patients without heart disease.

**Result:**

* T Statistic: **-2.1025**
* P Value: **0.0359**

**Conclusion:**
Since the p-value (0.0359) is less than 0.05, the null hypothesis is rejected. This indicates that cholesterol levels differ significantly between patients with heart disease and those without heart disease. Cholesterol appears to be an important clinical factor associated with heart disease risk.

---

### 2. Chi-Square Test

A Chi-Square Test was performed to examine the association between gender and heart disease occurrence.

**Result:**

* Chi-Square Statistic: **46.6239**
* P Value: **8.60 × 10⁻¹²**

**Conclusion:**
Since the p-value is significantly lower than 0.05, the null hypothesis is rejected. There is a statistically significant association between gender and heart disease occurrence. Gender may therefore be considered an important demographic factor in heart disease prediction.

---

### 3. ANOVA Test

ANOVA was applied to compare age distributions across different chest pain categories.

**Result:**

* F Statistic: **6.8134**
* P Value: **0.000160**

**Conclusion:**
Since the p-value is less than 0.05, the null hypothesis is rejected. This indicates that the average age differs significantly across chest pain categories. Age may influence the type or severity of chest pain experienced by patients.
